```python
import pandas as pd
import os, glob
import csv
  
demo = pd.concat(map(pd.read_csv, glob.glob(os.path.join('merge_datagov/Demography', "demo*.csv"))))
```

```python
demo = demo.fillna(demo.mean())
```

```python
demo.drop('Country/ States/ Union Territories Name', axis=1)
```

```python
demo.drop('Category', axis=1)
```

```python
max = demo.describe().iloc[7:8,:]
min = demo.describe().iloc[3:4,:]
min
```

```python
for column in demo:
    if column == 'Country/ States/ Union Territories Name':
        continue
    if column == 'Category':
        continue
    else:
       #df[column] = (df[column] - min.iloc[0][column])/(max.iloc[0][column] - min.iloc[0][column])
        demo[column] = demo[column]/max.iloc[0][column]
```
