```python
import numpy as np
list1 = []
list2 = []
for key in regions:
    if key == 'Andhra Pradesh':
        continue
    if key == 'West Bengal':
        continue
    if key == 'A & N Islands':
        continue
    if key == 'D & N Haveli':
        continue
    if key == 'Daman & Diu':
        continue
    if key == 'Lakshadweep':
        continue
    if key == 'NCT of Delhi':
        continue
    else:    
        list1.append(np.abs(eco["All_India GDP"] - eco[key]).sum())
        list2.append(key)
```
